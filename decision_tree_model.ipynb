```python
import pandas as pd
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('cleaned_data.csv')
```

```python
#spiting dataset for Analysis
#preparing featrure and target Variable

x = df.drop(columns=['attrition'])
y = df['attrition']
```

```python
#coverting all categorical data to numerrical
x = pd.get_dummies(x, drop_first=True)
```

```python
#Spliting dataframe fot train tests
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=92)
#0.3 mean 70/30 split of dataframe and  92 is random number
```

```python
# configuring a decision tree model
dt_model = DecisionTreeClassifier(max_depth=5, class_weight='balanced', random_state=92)
dt_model.fit(x_train, y_train)

# Prediction
y_pred_dt = dt_model.predict(x_test)

# checking 'Accuracy'
print("Accuracy:", accuracy_score(y_test, y_pred_dt))

# Generating Report
print("\nClassification Report:\n", classification_report(y_test, y_pred_dt))

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_dt)
sns.heatmap(cm, annot=True, fmt='d', cmap='Purples', xticklabels=['Stayed', 'Left'], yticklabels=['Stayed', 'Left'])
plt.title("Decision Tree Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```
