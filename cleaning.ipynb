```python
import pandas as pd
```

```python
#importing dataset

df = pd.read_csv('test.csv')
df.head()
```

```python
#Sorting column name
df.columns = df.columns.str.lower().str.replace(' ','_')
df.head()
```

```python
#handling Null, Duplicates,  values
df.isnull().sum()          #No Null values
df.duplicated().sum()      #No Duplicates values
```

```python
#dataype inspection
df.dtypes
```

```python
#checking for Inconsistencies in dataframe
for col in df.select_dtypes(include='object').columns:
    print(f"{col}: {df[col].unique()}")
```

```python
#checking for uniquw values in Job Satisfaction
print(df['job_satisfaction'].unique())
```

```python
#Formatting everything
df['job_satisfaction'] = df['job_satisfaction'].str.strip().str.title()
```

```python
#reaaplying Mapping to Job Satishfaction
df['job_satisfaction'] = df['job_satisfaction'].map({
    'Low': 1, 'Medium': 2, 'High': 3, 'Very High': 4
})
```

```python
#Checking for Values
print(df['job_satisfaction'].unique())
```

```python
#final verification of Job Satisfaction
print(df['job_satisfaction'].isna().sum())

print(df['job_satisfaction'].unique())
```

```python
#Converting Strings to Numerical value for better Visualization column by column

#work_life_balance
df['work-life_balance'] = df['work-life_balance'].map({
    'Poor': 1,
    'Fair': 2,
    'Good': 3,
    'Excellent': 4
})

#performance_rating
df['performance_rating'] = df['performance_rating'].map({
    'Low': 1,
    'Below Average': 2,
    'Average': 3,
    'High': 4
})

#employee_recognition
df['employee_recognition'] = df['employee_recognition'].map({
    'Low': 1,
    'Medium': 2,
    'High': 3,
    'Very High':4
})

#company_reputation
df['company_reputation'] = df['company_reputation'].map({
    'Poor': 1,
    'Fair': 2,
    'Good': 3,
    'Excellent': 4
})
```

```python
#cehcking mapping in applied column

print(df['job_satisfaction'].unique())
print(df['work-life_balance'].unique())
print(df['performance_rating'].unique())
print(df['employee_recognition'].unique())
print(df['company_reputation'].unique())
```

```python
df.to_csv('cleaned_data.csv', index=False)
print("downloaded newly updated cs:")
```
